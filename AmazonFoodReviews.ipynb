                                                             Vader Sentiment Scoring
VADER (Valence Aware Dictionary for Sentiment Reasoning) is a model used for text sentiment analysis that is sensitive to both polarity (positive/negative), it assigns each word a positive, negative and neutral score.

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('Very Best')
```

```python
sia.polarity_scores(example)
```

Polarity Score on entire dataset

```python
result = {}

for i,row in tqdm(df.iterrows(),total= len(df)):
    text = row['Text']
    myid = row['Id']
    result[myid] = sia.polarity_scores(text) 
```

                                                        Roberta Pre Trained Model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
#We import the data set and model from transformers library
task='sentiment'
MODEL = f"cardiffnlp/twitter-roberta-base-{task}"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
print(example)
sia.polarity_scores(example)
```

```python
encoded_text = tokenizer(example,return_tensors='pt')
print(encoded_text)
```

```python
output = model(**encoded_text)
output
```

```python
scores = output[0][0].detach().numpy()
print(scores)
```

```python
scores = softmax(scores)
scores
```

```python
scores_dict ={
    'roberta_neg':scores[0],
    'roberta_neu':scores[1],
    'roberta_poa':scores[2],
}
print(scores_dict)
```

```python
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example,return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict ={
        'roberta_neg':scores[0],
        'roberta_neu':scores[1],
        'roberta_pos':scores[2],
    }
    return scores_dict
```

```python
result = {}

for i,row in tqdm(df.iterrows(),total= len(df)):
    try:
        text = row['Text']
        myid = row['Id']

        vader_result= sia.polarity_scores(text) 

        vader_result_rename = {}
        for key, value in vader_result.items():
            vader_result_rename[f'vader_{key}']= value

        roberta_result= polarity_scores_roberta(text)
        both = (vader_result_rename | roberta_result)
        result[myid] = both
    except RuntimeError:
        print(f'Breaking the code at {myid}')
```

```python
result
```

```python
result_df = pd.DataFrame(result).T
result_df = result_df.reset_index().rename(columns={'index':'Id'})
result_df = result_df.merge(df,how ='left')
result_df.head()
```

```python
result_df.columns
```

```python
sns.pairplot(data = result_df, 
    vars = ['vader_neg', 'vader_neu', 'vader_pos','roberta_neg', 'roberta_neu', 'roberta_pos'],
    hue='Score',
    palette='tab10')
```

```python
fig, axs = plt.subplots(1,3,figsize=(15,3))

sns.barplot(data = result_df, x ='Score', y= 'roberta_pos', ax = axs[0]).set_title('Roberta_Positive')
sns.barplot(data = result_df, x ='Score', y= 'roberta_neg', ax = axs[1]).set_title('Roberta_Negative')
sns.barplot(data = result_df, x ='Score', y= 'roberta_neu', ax = axs[2]).set_title('Roberta_neutral')

plt.tight_layout()
```
